### txt file IO

```python
import os

with open("test.ksh", mode="r", encoding="utf8") as ksh:
    metadata = ksh.readlines()
print(ksh.name)
# for element in metadata[0:20]:
#     print(*element.split("=", 1), end="") #"*" unpacks a list

title = metadata[0][7:].rstrip("\n")
bpm = float(metadata[7][2:])
offset = int(metadata[10][2:])
preview_point = int(metadata[13][3:])
preview_length = int(metadata[14][8:])
print(title, bpm, offset, preview_point, preview_length)


metadata[0] = "title=" + title + " x" + str(1.34) + "\n"
metadata[7] = "bpm=" + str(round((bpm * 1.34), 3)) + "\n"
metadata[10] = "o=" + str(round((offset / 1.34))) + "\n"
metadata[13] = "po=" + str(round(preview_point / 1.34)) + "\n"
metadata[14] = "plength=" + str(round(preview_length / 1.34)) + "\n"

if (os.path.exists("testOutput/" + title)):
    pass
else:
    outputdir = os.mkdir("testOutput/" + title)

complete_name = os.path.join(("testOutput/" + title), ksh.name)
with open(complete_name, mode = "w") as output:
    output.writelines(metadata)
```
